# Menghitung jarak data ke setiap pusat cluster

Import library yang dibutuhkan

```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import pairwise_distances
```

Import dataset

```python
df = pd.read_excel('kemiskinan di pulau jawa missing.xlsx')
```

Mengisi nilai NaN dengan nilai median untuk kolom tertentu

```python
columns_to_fill = ['Jumlah penduduk miskin (000)', 'Presentase penduduk miskin']
df[columns_to_fill] = df[columns_to_fill].fillna(df[columns_to_fill].median())
```

Menyimpan nama kota dalam DataFrame

```python
kota = df['Kabupaten/Kota']
```

Memilih fitur yang akan digunakan untuk clustering

```python
X = df[['Jumlah penduduk miskin (000)', 'Presentase penduduk miskin']]
```

Standarisasi fitur

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

Jumlah cluster yang diinginkan

```python
n_clusters = 4
```

Melakukan clustering dengan K-Means

```python
kmeans = KMeans(n_clusters=n_clusters, random_state=0)
clusters = kmeans.fit_predict(X_scaled)
```

Menambahkan kolom cluster ke data asli

```python
df['Cluster'] = clusters
```

Menghitung jarak data ke setiap pusat cluster

```python
distances = pairwise_distances(X_scaled, kmeans.cluster_centers_)
```

Membuat DataFrame dengan keterangan nama kota dan matriks jarak

```python
distance_df = pd.DataFrame(distances, columns=[f'Cluster_{i+1}' for i in range(n_clusters)])
distance_df.insert(0, 'Kabupaten/Kota', kota)
```

Menampilkan DataFrame dengan keterangan

Import library yang dibutuhkan

```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import pairwise_distances
```

Import dataset

```python
import pandas as pd
df = pd.read_excel('kemiskinan di pulau jawa missing.xlsx')
```

Mengisi nilai NaN dengan nilai median untuk kolom tertentu

```python
columns_to_fill = ['Jumlah penduduk miskin (000)', 'Presentase penduduk miskin']
df[columns_to_fill] = df[columns_to_fill].fillna(df[columns_to_fill].median())
```

Memilih fitur yang akan digunakan untuk clustering

```python
X = df[['Jumlah penduduk miskin (000)', 'Presentase penduduk miskin']]
```

Standarisasi fitur

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

Jumlah cluster yang diinginkan

```python
n_clusters = 4
```

Melakukan clustering dengan K-Means

```python
kmeans = KMeans(n_clusters=n_clusters, random_state=0)
clusters = kmeans.fit_predict(X_scaled)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score

import pandas as pd
df = pd.read_excel('kemiskinan di pulau jawa missing.xlsx')

# Mengisi nilai NaN dengan nilai median untuk kolom tertentu
columns_to_fill = ['Jumlah penduduk miskin (000)', 'Presentase penduduk miskin']
df[columns_to_fill] = df[columns_to_fill].fillna(df[columns_to_fill].median())

# Memilih fitur yang akan digunakan untuk clustering
X = df[['Jumlah penduduk miskin (000)', 'Presentase penduduk miskin']]

# Standarisasi fitur
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Range jumlah cluster yang akan diuji
range_clusters = range(1, 11)  # Misalnya, mencoba dari 2 hingga 10 cluster

# Menyimpan Silhouette Score untuk setiap jumlah cluster
silhouette_scores = []

for n_clusters in range_clusters:
    kmeans = KMeans(n_clusters=n_clusters, random_state=0)
    cluster_labels = kmeans.fit_predict(X_scaled)
    silhouette_avg = silhouette_score(X_scaled, cluster_labels)
    silhouette_scores.append(silhouette_avg)

# Plot grafik Silhouette Score
plt.figure(figsize=(8, 6))
plt.plot(range_clusters, silhouette_scores, marker='o', linestyle='-')
plt.xlabel('Jumlah Cluster')
plt.ylabel('Silhouette Score')
plt.title('Metode Silhouette Score untuk Menentukan Jumlah Cluster Optimal')
plt.grid(True)
plt.show()
```
